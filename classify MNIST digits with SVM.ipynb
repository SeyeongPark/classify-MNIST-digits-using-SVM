```python
import tensorflow as tf

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

x_train_rs = x_train.reshape(60000, 28 * 28)
```

```python
x_test_rs = x_test.reshape(10000, 28 * 28)
```

```python
import pandas as pd

x_train_rs_df = pd.DataFrame(x_train_rs)
x_train_rs_df
```

```python
x_test_rs_df = pd.DataFrame(x_train_rs)
x_test_rs_df
```

```python
# only 20 rows will be used
x_train_rs_df.drop(x_train_rs_df.index[20 : ], axis='index', inplace=True) 
x_train_rs_df
```

```python
y_train_ser = pd.Series(y_train[ : 20 ])  #elements from whatever up to index 19 
```

```python
x_test_rs_df = pd.DataFrame(x_test_rs)
x_test_rs_df
```

```python
# only 5 rows will be used

x_test_rs_df.drop(x_test_rs_df.index[5 : ], axis='index', inplace=True) 
x_test_rs_df
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(x_train_rs_df, y_train_ser)

#Predict the response for test dataset
y_pred = clf.predict(x_test_rs_df)
```

```python
y_test_ser = pd.Series( y_test[:5] )
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test_ser, y_pred))
```

```python
# Model Precision
print("Precision:",metrics.precision_score(y_test_ser, y_pred, average='micro'))

# Model Recall
print("Recall:",metrics.recall_score(y_test_ser, y_pred, average='micro'))
```
